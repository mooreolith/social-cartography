## jquery.fourd.js

### Dependencies

Depends on the above and on 

* fourd.js.

### Functionality:
This library provides four main functions, and a few utility functions:


The data structure you're operating on is called a **graph**, and with this library, you can 
* add a **vertex** (The little black boxes)
* remove vertices (plural of vertex)
* add an **edge** between two vertices (the little lines between black boxes)
* remove an edge

Through these still fairly low level operations, one can already produce all manners of shapes. It's really cool!

#### Initialization
First, include a html element which you want to use to present the graph. 
```html
...
<div id="display">
</div>
...
```

Then, instantiate a fourd widget, turning the abovementined `div` into a `fourd`

```js
$("#display").fourd({width: 1000, height: 500, background: 0xffffff});
```

#### Vertex add_vertex(*VertexOptions)
This widget method adds a vertex to the graph, optionally supplying
* an integer for `size`
* a string containing a path to a publicly available square jpeg for `texture`.

All wrapped neatly in an object. 
```js
var vertex_0 = $('#display').fourd('add_vertex', {vertex_options});
```
Saving the vertex via `var` or `let` allows us to proceed with the next couple of commands. 

#### remove_vertex(Vertex)
Use this method to get rid of a vertex on the graph. Any edges connected to this graph also disappear. 

```js
$('#display').fourd('remove_vertex', vertex_0);
```

#### add_edge(Vertex, Vertex, *{EdgeOptions})
But first, you'll have to add one. Just supply two vertices as arguments. 

```js
var edge = $('#display').fourd('add_edge', vertex_0, vertex_1);
```

#### remove_edge(Edge)
Pretty self explanatory. 
```js
$('#display').fourd('remove_edge', edge);
```

#### clear()
For when you want to start over...
```js
$('#display').fourd('clear');
```

`*` means the argument is optionsl. 

### An easier way. 
You can get access to the widget's underlying fourd object by calling
```js
var fourd = $('#disolay').fourd('underlying_object');
```

This will provide you with a hole slew of options, some of which are advanced and only determined by trial and error (so there's plenty of room for improvement).

Here is some of the basic functionality described above, done with the javascript object returned by .fourd('underlying_object').

* ```js
var vertex = fourd.graph.add_vertex({VertexOptions});
```

* ```js
fourd.graph.remove_vertex(vertex);
```

* ```js
var edge = fourd.graph.add_edge(vertex, another_vertex);
```

* ```js
fourd.graph.remove_edge(edge);
```
